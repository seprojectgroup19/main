```python
from importlib import reload
import xgboost as xgb
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import ReadFromDB as RDB
from sklearn.model_selection import train_test_split
pd.options.display.max_columns = 999
pd.options.display.max_rows = 10
```

```python
def station_function(i):
    
    station = RDB.add_times(RDB.create_station_dictionary(i)[i])
    station_day = station[station.hour>6][station.hour<22]
    station_night = station.drop(station_day.index)
        
    weather = RDB.weather_times(RDB.read_weather(i))
    weather_day = weather[weather.hour>6][weather.hour<22]
    weather_night = weather.drop(weather_day.index)
    
    df_day = pd.merge_asof(station_day, weather_day, left_on=['last_updated'], right_on=['time'], tolerance=600, direction='nearest')
    df_day = df_day.dropna()
    df_day = df_day.reset_index()
    df_day.drop('index', axis=1, inplace=True)
    
    df_night = pd.merge_asof(station_night, weather_night, left_on=['last_updated'], right_on=['time'], tolerance=600, direction='nearest')
    df_night = df_night.dropna()
    df_night = df_night.reset_index()
    df_night.drop('index', axis=1, inplace=True)
    
    days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    weather_icons = ['partly-cloudy-day',
                     'partly-cloudy-night',
                     'clear-night',
                     'clear-day',
                     'fog',
                     'wind',
                     'cloudy',
                     'rain'
                    ]
    weather_cols = ['apparentTemperature',
                    'cloudCover',
                    'dewPoint',
                    'humidity',
                    'precipIntensity',
                    'precipProbability',
                    'pressure',
                    'temperature',
                    'windBearing',
                    'windGust',
                    'windSpeed',
                    'uvIndex',
                    'visibility'
                    ]
    
    for day in days:
        df_day[day] = 0
        df_night[day] = 0
    
    for icon in weather_icons:
        df_day[icon] = 0
        df_night[icon] = 0
        
    for j in df_day.index:
        val = df_day.loc[j, 'day_x']
        for day in days:
            if val == day:
                df_day.loc[j, day] = 1
        
        icon_val = df_day.loc[j, 'icon']
        for icon in weather_icons:
            if icon_val == icon:
                df_day.loc[j, icon] = 1
                
    for j in df_night.index:
        val = df_night.loc[j, 'day_x']
        for day in days:
            if val == day:
                df_night.loc[j, day] = 1
        
        icon_val = df_night.loc[j, 'icon']
        for icon in weather_icons:
            if icon_val == icon:
                df_night.loc[j, icon] = 1
    
    data_day = pd.DataFrame((df_day.Mon, df_day.Tue, df_day.Wed, df_day.Thu, df_day.Fri, df_day.Sat, df_day.Sun, df_day.hour_x)).T
    data_night = pd.DataFrame((df_night.Mon, df_night.Tue, df_night.Wed, df_night.Thu, df_night.Fri, df_night.Sat, df_night.Sun, df_night.hour_x)).T
    
    for icon in weather_icons:
        data_day[icon] = df_day[icon]
        data_night[icon] = df_night[icon]
    
    for col in weather_cols:
        data_day[col] = df_day[col]
        data_night[col] = df_night[col]
    
    label_day = pd.DataFrame((df_day.available_bikes))
    label_night = pd.DataFrame((df_night.available_bikes))
    
    return [df_day, data_day, label_day, df_night, data_night, label_night]
```

```python
def split_data(data, label):
    x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.2, shuffle=False)
    dtrain = xgb.DMatrix(x_train, label=y_train)
    dtest = xgb.DMatrix(x_test, label=y_test)
    return dtrain, dtest

def create_model(train, test, param, num_round):
    model = xgb.train(param, train, num_round, evals=[(test, 'test')], verbose_eval=50, early_stopping_rounds=50)
    return model

def cross_validation(train, param, num_round, folds):
    cvd = xgb.cv(param, train, num_round, nfold=folds)
    return cvd

def plot_errors(model , predictors, labels, station):
    
    inputs = xgb.DMatrix(predictors)
    predictions = model.predict(inputs)
    
    for i, pred in enumerate(predictions):
        if pred<0:
            predictions[i] = 0
        elif pred>station.bike_stands.max():
            prediction[i] = station.bike_stands.max()
    
    errors = list()
    
    for i, pred in enumerate(predictions):
        errors.append(pred-labels.loc[i])
        
    errors_sq = [i**2 for i in errors]
    
    plt.figure(figsize=(30,10))
    plt.plot(predictions, 'r-', alpha=0.75, label='Prediction')
    plt.plot(labels, alpha=0.75, label='Observation')
    plt.grid(True, which='both')
    plt.title('Predictions vs. Observations', fontsize=30)
    plt.ylabel('Available Bikes', fontsize=20)
    plt.legend(loc='best')
    
    plt.figure(figsize=(30,10))
    plt.plot(errors, alpha=0.75, label='Error')
    plt.grid(True, which='both')
    plt.title('Prediction Error, RMSE = {}'.format(np.round(np.sqrt(np.mean(errors_sq)),3)), fontsize=30)
    plt.ylabel('Error', fontsize=20)
```

```python
%%time
[station_day, data_day, label_day, station_night, data_night, label_night] = station_function(5)
```
